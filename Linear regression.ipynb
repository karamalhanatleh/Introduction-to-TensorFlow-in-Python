# Linear regression


```python
#import 
import numpy as np
import pandas as pd

import tensorflow as tf
```

```python
#Load data
housing = pd.read_csv("../data/kc_house_data.csv")
#head data
housing.head()
```

```python
#define the target and features

price= np.array(housing['price'] , np.float32)
size= np.array(housing['sqft_living'] , np.float32)
```

```python
#define intercept and slope
intercept = tf.Variable(0.1 , np.float32)
slope = tf.Variable(0.1 ,np.float32)
```

```python
#definne Linear redression model
def linear_regression(intercept , slope , features=size):
    return intercept+features*slope
```

```python
#compute predicted values and loss
def loss_function(intercept ,  slope , target=price , features=size):
    predictions=linear_regression(intercept , slope) 
    return tf.keras.losses.mse(target , predictions)
```

```python
#define an optimization operator
opt =tf.keras.optimizers.Adam()
```
